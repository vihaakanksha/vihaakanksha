## Histogram


This is a statistical plot (different from a bar plot), which shows the frequency of the values on the x axis, and not just categorical values.

Difference betwen histogram and bar charts: In most bar charts there will be space between bars, but most histograms won't.

```python
#generate data using the normal distribution
age=np.random.normal(20,1.5,1000) #20 is the mean, 1.5 is the std, 1000 is the total data points
print(age)
```

```python
#plot histogram based on the ages
plt.hist(age)
plt.show()
```

```python
#we can customize the bins manually
plt.hist(age,bins=2)
plt.show()
```

```python
#you can also pass a collection of bins like this:
plt.hist(age,bins=[age.min(),20,22,23,age.max()])
plt.show()
```

```python
plt.hist(age,bins=[age.min(),20,22,23,age.max()],edgecolor='green')
plt.show()
```

```python
#to plot cumulative histogram
plt.hist(age,bins=20, cumulative=True)
plt.show()
```

## Pie chart

Can be used for plotting, when you have independent categories, which do not necessarily overlap. This chart can be used to show the ratio of the data points.

```python
lang = ['C++',"C#","Python","Java","Go"]
votes = [20,50,140,1,45]
```

```python
plt.pie(votes, labels=lang)
plt.show()
#pie chart not optimal when the labels overlap.
```

```python
#for customization in pie chart, it is possible to explode certain pieces of the chart in python.
#Now we want to explode the "C++".
#We see that java is the 0th index.
#hence we create an array where the 0th index has a big value, as follows:


#these are the arrays we are using
lang = ['C++',"C#","Python","Java","Go"]
votes = [20,50,140,1,45]

#now we create the explodes array:
explodes=[0.5,0,0,0,0]

#So the 0.5 having the same index as the array locations we want to explode, we can now make our plot.
plt.pie(votes, labels=lang, explode=explodes)
plt.show()
```

```python
#writing % inside or outside of pie chart
plt.pie(votes, labels=lang, explode=explodes, autopct="%.2f%%") 
#you specify the autopct in here..."%here you write%".
#.2f defines the value until 2 floating points
plt.show()
```

```python
#specifying where the % value must be displayed
plt.pie(votes, labels=lang, explode=explodes, autopct="%.2f%%", pctdistance=1.3) 
plt.show()
```

```python
#if you want tospecify from where the pie chart must start you can do thiis:
plt.pie(votes, labels=lang, explode=explodes, autopct="%.2f%%", startangle=90)
plt.show()
```

## Box plot

Box plot is a statistical plot showing you the different quartiles, median, maximum, minimum and the outliers.

```python
#create a normal distribution
heights = np.random.normal(172,8,300)
print(heights)
```

```python
#plotting a boxplot
plt.boxplot(heights)
plt.show()
```

```python
#A better understanding of how boxplot works
first=np.linspace(0,10,25)
second=np.linspace(10,200,25)
third=np.linspace(200,210,25)
fourth=np.linspace(210,230,25)

data=np.concatenate((first, second, third, fourth))
```

```python
#plotting boxplot
plt.boxplot(data)
plt.show()

#we see that the big range is not because of more number of values, but because of bigger values.
```

## Plot Customization

```python
#Create some arrays
years=[2014,2015,2016,2017,2018,2019,2020,2021]
income=[55,56,62,61,72,72,73,75]
```

```python
#plot a line chart
plt.plot(years,income)
plt.title("Income of John (in $)")
plt.xlabel("Year")
plt.ylabel("Yearly Income in $")
plt.show()
```

### Some info on ticks:
Ticks are data point symbols that appear on axes.
Syntax is: matplotlib.pyplot.xticks(ticks=None, labels=None, ** kwargs)  
           
   where: 
           
   Ticks: This argument contains a list of xtick positions.
           
   Labels: This parameter comprises labels that will be placed at the specified tick places. It is also an optional parameter.
           
   ** kwargs: This parameter controls the appearance of the labels using Text attributes.

```python
# In our case the values on the y axis are $ in year. So they must have "k" after the number. Like, 55k, 57k etc.
#To do this, we do the following:
#1. define an array, which has values that contain all the yearly income on the y axis
#2. place this defined array in the labels parameter of the ticks function
years=[2014,2015,2016,2017,2018,2019,2020,2021]
income=[55,56,62,61,72,72,73,75]
income_ticks=list(range(50,81,2))

#plot a line chart
plt.plot(years,income)
plt.title("Income of John (in $)")
plt.xlabel("Year")
plt.ylabel("Yearly Income in $")
plt.yticks(income_ticks, [f"${x}k" for x in income_ticks])
plt.show()
```

```python
#Setting font sizes
#plot a line chart
plt.plot(years,income)
plt.title("Income of John (in $)",fontsize=30, fontname="Gadugi")

#you can specify the font size and name for xlabel and ylabel as well
plt.xlabel("Year")
plt.ylabel("Yearly Income in $")
plt.yticks(income_ticks, [f"${x}k" for x in income_ticks])
plt.show()
```

## Radar chart

Radar chart, also called as Spider chart or Web chart is a graphical method used for comparing multiple quantitative variables. It is a two dimensional polar visualization.
It is designed to show similarities, differences, and outliers in data.
The spokes in the radar are called radii. On these spokes, the values for the data are mapped.

Interpreting radar chart video for reference: https://www.youtube.com/watch?v=lpmFWXYf4NE

### When to use a radar chart?
1. There are multivariate observations
2. There is an arbitrary number of variables
3. You need to identify outliers
4. You need to make comparisons across products or services
5. Data sets are small or moderately sized

Advantage of radar chart: outliers and similarities are easily visible

### Disadvantage of radar chart
1. Hard to quantify Radii length by just seeing
2. Radar charts can distort data
3. Radar charts can create connections where there is none
4. Radar charts can cause occlusion and confusion when data is more

```python
style.use('ggplot')
```

```python
# marks obtained by 2 students- Alice and Bob in five subjects 
subjects=['PHY','CHEM','BIO','MATH', 'ECO']
alice=[60,40,68,94,27]
bob=[81,30,75,37,46]
```

```python
#matplotlib uses radians instead of angles

angles=np.linspace(0,2*np.pi,len(subjects), endpoint=False)  #why linspace?
#endpoint=True means that the last sample is the last value 

print(angles)
```

```python
#making 1st value as the last value
angles=np.concatenate((angles,[angles[0]]))
print(angles)
```

```python
#to maintain consistency, the 1st value should be appended to all the arrays
subjects.append(subjects[0])
alice.append(alice[0])
bob.append(bob[0])
```

```python
#plotting
fig=plt.figure(figsize=(6,6))  #creates a figure
ax=fig.add_subplot(polar=True)  #we are dealing with angles
ax.plot(angles,alice)
plt.show()
```

```python
#Customize for more info
fig=plt.figure(figsize=(6,6))
ax=fig.add_subplot(polar=True)

ax.plot(angles,alice, 'o--', color='g', label='Alice')

plt.legend()
plt.show()
```

```python
fig=plt.figure(figsize=(6,6))
ax=fig.add_subplot(polar=True)

ax.plot(angles,alice, 'o--', color='g', label='Alice')

#fill plot
ax.fill(angles, alice, alpha=0.25, color='g')

plt.legend()
plt.show()
```

```python
fig=plt.figure(figsize=(6,6))
ax=fig.add_subplot(polar=True)

ax.plot(angles,alice, 'o--', color='g', label='Alice')

#fill plot
ax.fill(angles, alice, alpha=0.25, color='g')

#Add labels
ax.set_thetagrids(angles * 180/np.pi, subjects)

plt.grid(True)
plt.tight_layout()

plt.legend()
plt.show()
```

```python
fig=plt.figure(figsize=(6,6))
ax=fig.add_subplot(polar=True)

ax.plot(angles,alice, 'o--', color='g', label='Alice')


#fill plot
ax.fill(angles, alice, alpha=0.25, color='g')
#Add labels
ax.set_thetagrids(angles * 180/np.pi, subjects)
plt.grid(True)
plt.tight_layout()
plt.legend()
plt.show()
```

```python
#Comparing Alice vs Bob 

fig=plt.figure(figsize=(6,6))
ax=fig.add_subplot(111, polar=True)
#Alice Plot
ax.plot(angles,alice, 'o-', color='g', linewidth=1, label='Alice')
ax.fill(angles, alice, alpha=0.25, color='g')
#Bob Plot
ax.plot(angles,bob, 'o-', color='orange', linewidth=1, label='Bob')
ax.fill(angles, bob, alpha=0.25, color='orange')
ax.set_thetagrids(angles * 180/np.pi, subjects)
plt.grid(True)
plt.tight_layout()
plt.legend()
plt.show()
```
